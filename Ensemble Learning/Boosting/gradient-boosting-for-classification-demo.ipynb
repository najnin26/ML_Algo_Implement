```python
import pandas as pd
import numpy as np
import random

df = pd.DataFrame(
                  [
                      [6.82, 118, 0],
                      [6.36, 125, 1],
                      [5.39, 99, 1],
                      [5.50, 106, 1],
                      [6.39, 148, 0],
                      [9.13, 148, 1],
                      [7.17, 147, 1],
                      [7.72, 72, 0]
                   ], columns = ['cgpa','iq','is_placed']
)

df
```

```python
# initial prediction

df['pre1(log-odds)'] = np.log(5/3)
df
```

```python
# convert log odds to probability
df['pre1(probability)'] = 1/(1+np.exp(-np.log(5/3)))
df
```

```python
# calculating residual for stage 1
df['res1'] = df['is_placed'] - df['pre1(probability)']
df
```
